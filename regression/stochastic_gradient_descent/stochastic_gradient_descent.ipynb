```python
import pandas as pd, math
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score
from sklearn.datasets import load_boston

dataset = load_boston()
X = pd.DataFrame(dataset.data, columns=dataset.feature_names)
y = pd.DataFrame(dataset.target, columns=['PRICE'])
df = pd.concat(objs=[X,y], axis=1).drop_duplicates().dropna()

df.head(10)
```

```python

df.plot(kind='scatter', x = 'LSTAT', y = 'PRICE')
plt.show()

fig, (hist1, hist2) = plt.subplots(1,2)
df['LSTAT'].plot(kind='hist', ax=hist1, title = 'LSTAT Histogram', xlabel= 'LSTAT')
df['PRICE'].plot(kind='hist', ax=hist2, title = 'PRICE Histogram', xlabel = 'PRICE')
plt.show()
```

```python
# Regression Model
X = df[['LSTAT']]
y = df['PRICE']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=0)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)

model = Ridge(alpha= 1.0, solver='auto', max_iter= 1000)
model.fit(X_train,y_train)

y_pre = model.predict(X_test)

print(f'Score: {r2_score(y_test, y_pre)}')

plt.scatter(X_train, y_train, c='b', marker='o', label = 'train')
plt.scatter(X_test, y_test, c='r', marker='o', label = 'test')
plt.plot(X_test, y_pre, c='g', label = 'model')
plt.xlabel('LSTAT')
plt.ylabel('PRICE')
plt.title('LSTAT vs PRICE')
plt.legend(loc='lower right')
plt.show()
```
